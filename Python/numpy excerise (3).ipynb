
#### Exercise 8:write a program that accepts a sentence and find value counts of letter and digits.Input value will be supplied to program

```python
str=input("Enter the string:")
dict={"letter":0,"digit":0}
for i in str:
    if(i.isdigit()):
        dict["digit"]+=1
    elif(i.isalpha()):
        dict["letter"]+=1
    
print('Letters: ',dict['letter'])
print('Digits: ',dict['digit'])
```

#### Exercise 9:Read a text file poem.txt using file open() and plot barplot/lineplot of words and its counts.

```python
import pandas as pd



def word_count(str):
    d = {}
    words = str.split()
    for word in words:
        if word not in d:
            d[word] = 1
        else:
            d[word] += 1
    return d



file = open('poem.txt','r')
sentence = file.read()
pd.Series(word_count(sentence)).plot.bar()
```
